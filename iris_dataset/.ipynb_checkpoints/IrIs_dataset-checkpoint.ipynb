# 2.資料前處裡
    @Missing data
        將缺的資料補值，因new_type的資料很少，所以補上new_type此欄的平均值

```python
#找出有空值的欄位
train[train['花萼長度'].isnull() | train['花萼寬度'].isnull()]
```

可發現，Iris-new_type只有三筆資料，其中兩筆資料有缺失

```python
#將空值的欄位填入Iris-new_type的平均值
train['花萼長度'] = train['花萼長度'].fillna(train['花萼長度'][train['屬種'] == 'Iris-new_type'].mean())
train['花萼寬度'] = train['花萼寬度'].fillna(train['花萼寬度'][train['屬種'] == 'Iris-new_type'].mean())
train[train['屬種'] == 'Iris-new_type']
```

```python
#check資料
train.info()
```

    資料切分


```python
from sklearn.model_selection import train_test_split
```

```python
# 目標變數
df_x= train[['花萼長度','花萼寬度','花瓣長度','花瓣寬度']]
# 預測變數
df_y=train[['type']]

#train_test_split 數據隨機切分
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size = 0.3)
```

```python
x_train #train_test_split
```

# 3.選擇及建立模型
    有labeled data，且資料小於100K ==>> Linear SVC 
    不是Text Data(文字資料) ==>> KNeighborsClassifier

```python
from IPython.display import Image
Image(filename='sklearnModelImage.png')
```

```python
# LinearSVC
from sklearn.svm import LinearSVC
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

LinearSVC_model = LinearSVC()
LinearSVC_model.fit(x_train,y_train.values.ravel())
```

```python
KNeighborsClassifier_model = KNeighborsClassifier()
KNeighborsClassifier_model.fit(x_train, y_train.values.ravel())
```

# 4.分析結果及修正模型

```python
#LinearSVC
#x_test 預測結果
print(LinearSVC_model.predict(x_test))
#準確率
print(LinearSVC_model.score(x_test,y_test))
```

```python
#KNeighborsClassifier
#x_test 預測結果
print(KNeighborsClassifier_model.predict(x_test))
#準確率
print(KNeighborsClassifier_model.score(x_test,y_test))
```

    結果: 使用KNeighborsClassifier model的準確率比較高

# 5.結果呈現
    使用KNeighborsClassifier model

```python
#載入test資料
predict_test = KNeighborsClassifier_model.predict(test[['花萼長度','花萼寬度','花瓣長度','花瓣寬度']])
```

```python
submission = pd.DataFrame({'id':test.id , 'type':predict_test} )
submission = submission[['id','type']]
submission.to_csv('submission.csv', index=False)
```
