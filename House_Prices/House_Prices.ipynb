# Missing Data

```python
#allData 先把train的資料與test的資料合併起來做處理，資料才會有一致性。
allData = pd.concat((train.loc[:,'MSSubClass':'SaleCondition'], test.loc[:,'MSSubClass':'SaleCondition']), axis=0,ignore_index=True)

#DataFrame形狀
print(allData.shape)

#查看全部欄位數值型態
print(allData.dtypes.value_counts())

#查看表格資訊
print(train.info())
```

```python
#MSSubClass在資料的定義上不是數值，將MSSubClass的值轉換成Str
allData['MSSubClass'] = allData['MSSubClass'].apply(str)
```

```python
#找出null資料，並排序
MissingData = allData.isnull().sum() 
MissingData = MissingData[MissingData>0] 
MissingData.sort_values(ascending=False) 
```

```python
#將MissingData補資料

#PoolQC,MiscFeature,Alley,Fence,FireplaceQu的NA意思是None的意思，先將其填上None

#PoolQC: Pool quality / NA No Pool
allData['PoolQC'] = allData['PoolQC'].fillna('None')

#MiscFeature: Miscellaneous feature not covered in other categories /NA None
allData['MiscFeature'] = allData['MiscFeature'].fillna('None')

#Alley: Type of alley access to property / NA 	No alley access
allData['Alley'] = allData['Alley'].fillna('None')

#Fence: Fence quality / NA	No Fence
allData['Fence'] = allData['Fence'].fillna('None')

#FireplaceQu: Fireplace quality / NA	No Fireplace
allData['FireplaceQu'] = allData['FireplaceQu'].fillna('None')

#LotFrontage: Linear feet of street connected to property
#相同的Neighborhood可能有相同的LotFrontage，所以先用Neighborhood找出其LotFrontage的中位數填入
allData['LotFrontage'] = allData.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

#Garage系列:GarageType NA	No Garage，代表也沒有GarageQual,GarageFinish,GarageYrBlt,GarageCond,GarageCars,GarageArea
allData['GarageQual'] = allData['GarageQual'].fillna('None')
allData['GarageFinish'] = allData['GarageFinish'].fillna('None')
allData['GarageCond'] = allData['GarageCond'].fillna('None')
allData['GarageType'] = allData['GarageType'].fillna('None')

allData['GarageYrBlt'] = allData['GarageYrBlt'].fillna(0)
allData['GarageCars'] = allData['GarageCars'].fillna(0)
allData['GarageArea'] = allData['GarageArea'].fillna(0)

#Bsmt系列:BsmtExposure,BsmtCond,BsmtQual,BsmtFinType2,BsmtFinType1,BsmtHalfBath,BsmtFullBath,BsmtFinSF2,BsmtUnfSF,BsmtFinSF1
#也是一樣，沒有帶None或0
for col in ('BsmtExposure','BsmtCond','BsmtQual','BsmtFinType2','BsmtFinType1'):
    allData[col] = allData[col].fillna('None')
    
for col in ('BsmtHalfBath','BsmtFullBath','BsmtUnfSF','BsmtFinSF2','BsmtFinSF1','TotalBsmtSF'):
    allData[col] = allData[col].fillna(0)

#MasVnrType: Masonry veneer type / None	None
allData['MasVnrType'] = allData['MasVnrType'].fillna('None')
#MasVnrArea: 沒有MasVnrType就沒有MasVnrArea
allData['MasVnrArea'] = allData['MasVnrArea'].fillna(0)

#MSZoning:Identifies the general zoning classification of the sale.
#填上眾數
allData['MSZoning'] = allData['MSZoning'].fillna(allData['MSZoning'].mode()[0])

#Functional: Home functionality (Assume typical unless deductions are warranted)
allData['Functional'] = allData['Functional'].fillna('Typ')

#Utilities: Type of utilities available
allData['Utilities'] = allData['Utilities'].fillna(allData['Utilities'].mode()[0])

#SaleType: Type of sale
allData['SaleType'] = allData['SaleType'].fillna(allData['SaleType'].mode()[0])

#KitchenQual: Kitchen quality
allData['KitchenQual'] = allData['KitchenQual'].fillna(allData['KitchenQual'].mode()[0])

#Electrical: Electrical system
allData['Electrical'] = allData['Electrical'].fillna(allData['Electrical'].mode()[0])

#Exterior1st,Exterior2nd 
allData['Exterior1st'] = allData['Exterior1st'].fillna(allData['Exterior1st'].mode()[0])
allData['Exterior2nd'] = allData['Exterior2nd'].fillna(allData['Exterior2nd'].mode()[0])

#Utilities:在train的資料只有一筆NoSeWa，其他都是AllPub，無助於特徵訓練，將其drop掉
allData = allData.drop(['Utilities'], axis=1)

#check Null
MissingData = allData.isnull().sum() 
MissingData = MissingData[MissingData>0] 
MissingData.sort_values(ascending=False) 
```

# 分布分析 (對數轉換)
    對數轉換的目的:
    為了減少數值差距過大及離群值而帶來的影響，從絕對差異轉變為相對差異，使得因變量的方差(離散程度)穩定
    
    skewness(偏度):統計數據分佈偏斜方向和程度的度量，是統計數據分佈非對稱程度的數字特徵。
    偏度定義中包括正態分佈（偏度=0），右偏分佈（也叫正偏分佈，其偏度>0），左偏分佈（也叫負偏分佈，其偏度<0）。

    1.target的分布 (SalePrice)

```python
from scipy import stats #統計函數庫

fig = plt.figure(figsize=(12,5))
plt.subplot(121)
sns.distplot(train_new.SalePrice, fit=stats.norm) #stats.norm 正態分佈
plt.subplot(122)
res = stats.probplot(train_new.SalePrice, plot=plt)
```

# Model

```python
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.svm import SVR, LinearSVR # LinearSVR與使用kernal為linear的SVR是等價的
from sklearn.linear_model import ElasticNet, BayesianRidge
from sklearn.kernel_ridge import KernelRidge
```

## 評價函數 rmse

```python
def rmse_cv(model,X,y):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=5))
    return rmse
```

## Regression Model

```python
models = [LinearRegression(),Ridge(),Lasso(alpha=0.01,max_iter=10000),SVR(),LinearSVR(),
          ElasticNet(alpha=0.001,max_iter=10000),BayesianRidge(),
          KernelRidge(alpha=0.6, kernel='polynomial', degree=2, coef0=2.5)]

names = ['LR', 'Ridge', 'Lasso', 'SVR', 'LinSVR', 'Ela','Bay','Ker']
for name, model in zip(names, models):
    score = rmse_cv(model, df_x_train, df_y)
    print('{}: {:.6f}, {:.4f}'.format(name,score.mean(),score.std()))
```

    LR Ridge Ela Bay Lasso 這幾個的rmse數值是比較少的，再將這幾個model以不同參數來看看

```python
class grid():
    def __init__(self,model):
        self.model = model
    
    def grid_get(self,X,y,param_grid):
        # 設定greedy search參數，能夠依照param_dict的預設值進行訓練，評價函數為mse
        grid_search = GridSearchCV(self.model,param_grid,cv=5, scoring='neg_mean_squared_error')
        grid_search.fit(X,y)
        print(grid_search.best_params_, np.sqrt(-grid_search.best_score_))
        grid_search.cv_results_['mean_test_score'] = np.sqrt(-grid_search.cv_results_['mean_test_score'])
        print(pd.DataFrame(grid_search.cv_results_)[['params','mean_test_score','std_test_score']])
```

```python
#Lasso
grid(Lasso()).grid_get(df_x_train,df_y,{'alpha': [0.0004,0.0005,0.0007,0.0009],'max_iter':[10000]})
```

```python
#Ridge
grid(Ridge()).grid_get(df_x_train,df_y,{'alpha':[0.001,0.1,1,10,35,40,45,50,55,60,65,70,80,90]})
```

```python
#SVR
grid(SVR()).grid_get(df_x_train,df_y,{'C':[5,7,11,13,15],'kernel':["rbf"],'gamma':[0.0003,0.0004],'epsilon':[0.008,0.009]})
```

```python
#KernelRidge
param_dict={'alpha':[0.2,0.3,0.4], 'kernel':["polynomial"], 'degree':[1,2],'coef0':[0.8,1]}
grid(KernelRidge()).grid_get(df_x_train,df_y,param_grid=param_dict)
```

```python
#ElasticNet
grid(ElasticNet()).grid_get(df_x_train,df_y,{'alpha':[0.0008,0.001,0.004,0.005],'l1_ratio':[0.08,0.1,0.3],'max_iter':[10000]})
```

```python
model = ElasticNet(alpha=0.0008,l1_ratio=0.3,max_iter=10000)
model.fit(df_x_train, df_y)
```

```python
predict = np.expm1(model.predict(df_x_test))
```
